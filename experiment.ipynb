```python
import numpy as np
import pickle, os
import matplotlib.pyplot as plt
```

```python
fn_lst = os.listdir('./output/')
fn_target = []
for fun in ['Sigmoid', 'ReLU', 'tanh']:
    for fn in fn_lst:
        if '2_lenet5' in fn and fun in fn and 'bs=64' in fn:
            fn_target.append(fn)
            
fn_target.remove('21-04-15-14-32-01_2_lenet5_tanh_fs=3_bs=64_epochs=30')
fn_target.remove('21-04-15-10-55-15_2_lenet5_tanh_fs=3_bs=64_epochs=30')
```

```python
d_loss = {}
for fn in fn_target:
    with open('./output/' + fn + '/Trainer.pkl', 'rb') as f:
        trainer = pickle.load(f)['trainer']
    s = fn.split('_')
    d_loss['_'.join(s[3:5])] = trainer.train_loss
    del trainer
```

```python
plt.figure(figsize=(18, 10))
for k, v in d_loss.items():
    plt.plot(v, label=k)
plt.ylim(3.75, 4.25)
plt.legend()
plt.show()
```
